## Import các thư viện cần thiết

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 
```

## Random hàm số

```python
x = 2 - 3 * np.random.normal(0, 1, 200)
y = np.random.normal(20)*x - np.random.normal(20)*(x**2) + np.random.normal(20)*(x**3) + np.random.normal(-3,3,200)
plt.scatter(x,y)
plt.show()
```

## Tìm kiếm hệ số a,b,c,d của hàm bậc 3

```python
def test(x,a,b,c,d): 
    return a*(x**3) + b*(x**2) + c*(x**2) + d
param, param_cov = curve_fit(test, x, y) 
param
```

## Biểu diễn giữa curve fit và hàm số ban đầu

```python
ans = param[0]*(x**3) + param[1]*(x**2) + param[2]*(x**2) + param[3]
plt.plot(x, y, 'o', color ='red', label ="data") 
plt.plot(x, ans, '--', color ='blue', label ="optimized data") 
plt.legend() 
plt.show() 
```

## Dùng zip() để sorted giá trị theo thứ tự từ thấp đến cao

```python
ans = param[0]*(x**3) + param[1]*(x**2) + param[2]*(x**2) + param[3]
plt.plot(x, y, 'o', color ='red', label ="data") 
new_x, new_ans = zip(*sorted(zip(x, ans)))
plt.plot(new_x, new_ans, '--', color ='blue', label ="optimized data") 
plt.legend() 
plt.show() 
```
